```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('copy.csv', index_col=0)
df.head()
```

```python
df = df.dropna()
```

```python
res = pd.DataFrame(df.restuarantID.value_counts())
```

```python
len(set(df['restuarantID'].values))
```

```python
res = res[res['restuarantID'] > 30]
```

```python
res['index1'] = res.index
res.columns = ['count', 'restuarantID']
res
```

```python
df_rest_final = df.merge(res,how = 'inner', on='restuarantID')
```

```python
df_rest_final = df_rest_final.drop(columns = ['count'])
```

```python
df_rest_final = df_rest_final.sample(frac=1).reset_index(drop=True)
```

```python
df_rest_final = df_rest_final.sort_values('userID')
```

```python
df_user = df_rest_final.copy()
```

```python
res = pd.DataFrame(df_user.userID.value_counts())
```

```python
res = res[res['userID'] > 30]
```

```python
res['index1'] = res.index
res.columns = ['count','userID']
```

```python
res
```

```python
df_user_final = df_user.merge(res,how = 'inner', on='userID')
```

```python
df_user_final = df_user_final.drop(columns=['count'])
```

```python
df_user_final.to_csv('userwisedata.csv', index = False)
```

```python
df_final = pd.read_csv('userwisedata.csv')
```

```python
df_res = df_final.sort_values('restuarantID')
```

```python
index = -1
prev = -1
for i,r in df_res.iterrows():
    if prev!=int(r['restuarantID']):
        index+=1
    prev = r['restuarantID']
    df_res.loc[i,'restuarantID']=index
```

```python
df_res.to_csv('df_res.csv', index=0)
df_res
```

```python
df_use = df_res.sort_values('userID')
```

```python
index = -1
prev = '-1'
for i,r in df_use.iterrows():
    if prev!=(r['userID']):
        index+=1
    prev = r['userID']
    df_use.loc[i,'userID']=index
```

```python
len(set(df_res['userID'].values))
```

```python
cnts = pd.DataFrame(df_use['userID'].value_counts())
```

```python
inds = np.arange(len(cnts.index))
```

```python
cnts['userInds'] = inds
```

```python
cnts['index1'] = cnts.index
cnts.columns = ['count', 'userInds','userID']
```

```python
cnts
```

```python
df_use2 = df_use.merge(cnts,how = 'inner', on='userID')
```

```python
len(set(df_use2['userInds'].values))
```

```python
df_use2
```

```python
df_clean = df_use2[['userInds','restuarantID','userRating']].sort_values('userInds')
```
